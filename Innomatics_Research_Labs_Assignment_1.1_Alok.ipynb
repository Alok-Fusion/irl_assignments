```python
##Problem 1: Running sum of 1D Array
```

```python
class Solution:
    def runningSum(self, nums):
        for i in range(1,len(nums)):
            nums[i]+=nums[i-1]
        return nums
sol = Solution()
print(sol.runningSum([1,2,3,4]))
```

```python
##Problem 2: Shuffle the Array
```

```python
class Solution:
    def shuffle(self, nums, n):
        resarr = []
        j = n
        for i in range(n):
            resarr.append(nums[i])
            resarr.append(nums[j+i])
        return resarr
sol = Solution()
nums,n = [3,5,6,2,7,2,8,2],4
print(sol.shuffle(nums,n))
```

```python
## Problem 3: Kids with greatest number of candies
```

```python
class Solution:
    def kidsWithCandies(self, candies, extraCandies):
        maxNumberOfCandies = max(candies)
        finalList = []

        for candy in candies:
            if candy + extraCandies >= maxNumberOfCandies:
                finalList.append(True)
            else:
                finalList.append(False)

        return finalList

sol = Solution()
candies, extra = [2,3,5,1,3],3
print(sol.kidsWithCandies(candies,extra))
```
