## Problem 2 Python if-else

```python
n = int(input())
    
if n % 2 != 0: 
    print("Weird")
elif n % 2 == 0 and 2 <= n <= 5: 
    print("Not Weird")
elif n % 2 == 0 and 6 <= n <= 20: 
    print("Weird")
elif n % 2 == 0 and n > 20: 
    print("Not Weird")
```

## Problem 6 Write a funtion

```python
def is_leap(year):
    leap = False
    if (year % 4 == 0 and year %100 != 0) or year %400 == 0 :
        leap = True
        return leap
    else : 
        leap = False
        return leap
    return leap

year = int(input())
print(is_leap(year))
```

## Problem 7 print() function

```python
n = int(input())
for i in range(1,n+1):
    if i < n:
        print(i,end="")
    else:
        print(i)
```
